```python
input_data['SampleDateUpdate'] = pd.to_datetime(input_data['SampleDate'], format='%m/%d/%Y %H:%M:%S %p').dt.strftime('%m/%d/%Y')
```

```python
input_data = input_data.dropna(axis=1, how='all')
```
